```python
#Importando as bibliotecas pandas e plotly.express para análise e visualização de dados em Python.


import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel("/content/acoes.xlsx",sheet_name="Principal")
```

```python
# usando o método .head(10) do pandas para exibir as primeiras 10 linhas do DataFrame df_principal, permitindo uma rápida visualização dos dados

df_principal.head(10)
```

```python
# Lendo a planilha "Total_de_acoes" do arquivo "acoes.xlsx" e armazenando os dados em df_total_acoes.

df_total_acoes = pd.read_excel("/content/acoes.xlsx",sheet_name="Total_de_acoes")
df_total_acoes
```

```python
# Lendo a planilha "Ticker" do arquivo "acoes.xlsx" e armazenando os dados em df_total_ticker e os exibindo.

df_total_ticker = pd.read_excel("/content/acoes.xlsx",sheet_name="Ticker")
df_total_ticker
```

```python
# Selecionando as colunas 'Ativo', 'Data', 'valor_final' e 'Var. Dia (%)' do DataFrame df_principal e criando uma cópia do DataFrame com essas colunas selecionadas.
df_principal = df_principal[['Ativo', 'Data','Último (R$)',	'Var. Dia (%)']].copy()

df_principal
```

```python
# Renomeando colunas e criando uma cópia do DataFrame principal
df_principal= df_principal.rename(columns={'Último (R$)':'valor_final', 'Var. Dia (%)':'var_dia_pct'}).copy()
df_principal
```

```python
#  realizando cálculos no DataFrame df_principal para criar duas novas colunas:
# var_pct: Calculando a porcentagem dividindo a coluna 'var_dia_pct' por 100.
# valor_inicial: Calculando o valor inicial dividindo a coluna 'valor_final' pelo resultado da fórmula (var_pct + 1), para realizar
# cálculo de variações percentuais e valores iniciais

df_principal['var_pct'] = df_principal ['var_dia_pct'] / 100

df_principal['valor_inicial'] = df_principal ['valor_final'] / (df_principal['var_pct'] +1)
df_principal
```

```python
# Aqui estou juntando as informações em uma unica planilha, buscando como se fosse um procv

df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python

df_principal
```

```python
# Calculando a variação em reais e adicionando uma nova coluna ao DataFrame principal
df_principal['Variacao_rs'] = (df_principal['valor_final'] - df_principal['valor_inicial'])*df_principal['Qtde. Teórica']
df_principal
```

```python
# Definindo o formato de exibição dos números decimais no DataFrame principal

pd.options.display.float_format = '{:.2f}'.format
df_principal
```

```python
# Convertendo a coluna 'Qtde. Teórica' para o tipo de dado inteiro (int) no DataFrame principal

df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
# Renomeando a coluna 'Qtde. Teórica' para 'Qtd_teorica' e criando uma cópia do DataFrame principal

df_principal = df_principal.rename(columns={'Qtde. Teórica':'Qtd_teorica'}).copy()
df_principal
```

```python
# Calculando o resultado com base na variação em reais e adicionando uma nova coluna ao DataFrame principal
df_principal['Resultado'] = df_principal['Variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
df_principal
```

```python
df_ticker = pd.read_excel("/content/acoes.xlsx", sheet_name="Ticker")
df_ticker
```

```python
# Mesclando os DataFrames df_principal e df_ticker com base nas colunas 'Ativo' e 'Ticker', respectivamente, usando um left join
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on='Nome', right_on='Empresa', how='left')
df_principal = df_principal.drop(columns=['Empresa'])
df_principal
```

```python
df_principal = df_principal.rename(columns={'Idade em anos':'Idade'}).copy()
df_principal
```

```python
# Criando uma nova coluna 'Cat_idade' com base na idade e adicionando ao DataFrame principal

df_principal['Cat_idade'] = df_principal['Idade'].apply(lambda x: 'Mais de 100' if x > 100 else ('Menos de 50' if x < 50 else 'Entre 50 e 100'))
df_principal
```

```python
# Calculando o maior valor
maior = df_principal['Variacao_rs'].max()

# Calculando o menor valor
menor = df_principal['Variacao_rs'].min()

# Calculando a média
media = df_principal['Variacao_rs'].mean()

# Calculando a média de quem subiu
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['Variacao_rs'].mean()

# Calculando a média de quem desceu
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['Variacao_rs'].mean()

# Imprimindo os resultados
print(f"Maior\tR$ {maior:,.2f}")
print(f"Menor\tR$ {menor:,.2f}")
print(f"Média\tR$ {media:,.2f}")
print(f"Média de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Média de quem desceu\tR$ {media_desceu:,.2f}")
```

```python
# Criando um novo DataFrame df_principal_subiu contendo apenas as linhas onde o resultado foi 'Subiu'
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

```python
# Agrupando o DataFrame df_principal_subiu por 'Segmento' e calculando a soma das variações em reais

df_analise_segmento = df_principal_subiu.groupby('Segmento')['Variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
# Agrupando o DataFrame df_principal por 'Resultado' e calculando a soma das variações em reais


df_analise_saldo = df_principal.groupby('Resultado')['Variacao_rs'].sum().reset_index()
df_analise_saldo
```

```python
# Criando um gráfico de barras para visualizar a variação em reais por resultado
fig = px.bar(df_analise_saldo, x='Resultado', y='Variacao_rs', text='Variacao_rs', title = 'Variação Reais por Resultado')
fig.show()
```

```python
# Atualizando o template de texto para exibir os valores com duas casas decimais no gráfico de barras

fig.update_traces(texttemplate='%{text:,.2f}')
fig.show()
```

```python
# Criando um gráfico de pizza para visualizar a variação em reais por segmento
figpizza = px.pie(df_analise_segmento, names='Segmento', values='Variacao_rs', title='Variação Reais por Segmento')
figpizza.show()
```

```python
# Agrupando o DataFrame df_principal por 'Cat_idade' e calculando a soma das variações em reais
df_analise_cat_idade = df_principal.groupby('Cat_idade')['Variacao_rs'].sum().reset_index()
df_analise_cat_idade
# Criando um gráfico de barras para visualizar a variação em reais por categoria de idade
figidade = px.bar(df_analise_cat_idade, x='Cat_idade', y='Variacao_rs', text='Variacao_rs', title='Variação Reais por Categoria de Idade')
figidade.show()
```
